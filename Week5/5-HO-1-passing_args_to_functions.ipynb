# Function Arguments

```python
def print_func(man1, man2, *args, **kwargs):
  print(f'man1: {man1}')
  print(f'man1: {man1}')

  for arg in args:
    print(f'arg: {arg}')

  for kwarg_key, kwarg_val in kwargs.items():
    print(f'{kwarg_key}: {kwarg_val}')
```

```python
print_func(10, 100, 1000, 1, x='a', y='e')
```
