```python
def add_list(list1, list2):
  if (len(list1) != len(list2)):
    raise ValueError('Lists are of unequal lengths')

  result_list = []
  for i,j in zip(list1, list2):
    result_list.append(i+j)

  return result_list
```

```python
add_list(c,b)
```
