```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def draw_pair_dice_sum(n_reps: int = 10):
  np.random.seed(123)
  x1 = 1 + np.random.choice(5, n_reps)
  np.random.seed(321)
  x2 = 1 + np.random.choice(5, n_reps)

  return x1 + x2
```

```python
# fix draws, increase repitions
# here, draws from the box [1,2,3,4,5,6] is 2
# repitition changes

n_reps = [10, 100, 1000, 10000]

for reps in n_reps:
  data = draw_pair_dice_sum(reps)
  bins, inv = np.unique(data, return_inverse=True)
  counts = np.bincount(inv)
  plt.bar(bins, counts, width= 1)
  plt.show()
```

```python
# function for draws:

def sum_n_draws(n_draws):
  sum = 0
  for i in range(n_draws):
    sum += (1 + np.random.choice(5, 1)) # draws has to be > 0, so 1+i

  return sum
```

```python
# let's fix repitions fixed at 100
# n_draws = 10
n_reps = 100
sum_list = []
n_draws = 10
for i in range(n_reps):
  sum_list.append(sum_n_draws(n_draws = n_draws))

#plot
bins, inv = np.unique(np.array(sum_list), return_inverse=True)
counts = np.bincount(inv)
plt.bar(bins, counts, width= 1)
```
