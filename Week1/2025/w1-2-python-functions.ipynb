# Python Functions

## How to define a function?

```python
def add(x, y):
  return x + y
```

```python
def power(x, y):
  return x ** y
```

```python
_ = power(11, 2)
```

## Pay attention to data types of the arguments

```python
add('a', 'b')
```

```python
try:
  power('a', 'b')
except TypeError as e:
  print(f"Caught an exception: {e}")
```

## Lambda functions

```python
# defining a function of function
square = lambda x: power(x, 2)
```

```python
square(11)
```

## Function arguments
1. positional & arbitrary positional
2. default
3. key word

### Positional arguments

```python

def greet(name, age):
    print(f"Hello, {name}! You are {age} years old.")


greet("Alice", 25)
greet(age=25, name="Alice") # alternate 1
greet("Alice", age = 25) # alternate 2

# greet(age=25, "Alice") # not all argument names explicit, order is important
```

### Positional args with default values

```python
# With default agrument values
def greet(name, age=25):
    print(f"Hello, {name}! You are {age} years old.")


greet("Alice")
greet("Alice", 10)
greet("Alice", age=10)
```

```python
# using keyword arguments
name_age_dict = {'name': 'Alice', 'age': 25}

greet(**name_age_dict)
```

### Arbitrary positional arguments

```python
def sum_numbers(*args):
    total = 0
    for num in args:
        total += num
    print(f"sum of supplied numbers: {total}")

sum_numbers(1,2,3,4)
sum_numbers(1,2)
```

### Kwargs

```python
# With default agrument values
def greet(name, age=25, **kwargs):
    print(f"Hello, {name}! You are {age} years old.")

    for key, val in kwargs.items():
        print(f"misc. attr {key}: {val}")

greet("Alice")
greet("Alice", city='Bengaluru', country='India')
```
