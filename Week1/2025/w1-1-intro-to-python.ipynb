# Control flow, loops

```python
# explain how if-else works

a = 4

if (a > 2):
  print(f'a > 2. value of a = {a}')
elif (a == 2):
  print(f'a == 2')
else:
  print(f'a < 2. value of a = {a}')
```

```python
# explain the format

for i in alist:
  print (f'An item in the list "alist": {i}')
```

```python
list(enumerate(alist))
```

```python
# get the index of each item
for i, item in enumerate(alist):
  idx = i + 1
  if (idx == 1):
    suffix = 'st'
  elif (idx == 2):
    suffix = 'nd'
  elif (idx == 3):
    suffix = 'rd'
  else:
    suffix = 'th'
  print (f'{idx}{suffix} item in "alist": {item}')

# find out what zip(list_1, list_2) does?
```
